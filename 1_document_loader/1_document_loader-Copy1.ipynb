# Introduction

The real estate industry generates large volumes of information including property listings, pricing details, location insights, amenities, legal documents, and market trends. Customers often spend significant time searching through multiple sources to find relevant property information.

This project introduces an **AI-powered Real Estate Assistant using Retrieval-Augmented Generation (RAG)** that combines semantic search, vector databases, and Large Language Models (LLMs) to provide accurate, context-aware, and intelligent responses to real estate-related questions.

The system retrieves relevant information from a property knowledge base and generates human-like responses grounded in factual data, reducing hallucinations and improving answer reliability.

# 🌐 Step 1: Document Loading using URL Loader

The first stage of the Retrieval-Augmented Generation (RAG) pipeline involves collecting data from external sources. In this project, LangChain's **UnstructuredURLLoader** is used to load and extract content from web pages containing real estate information.

The loader is imported from LangChain's document loader module and initialized with one or more URLs. These URLs may contain:

- Property listings
- Real estate news articles
- Market reports
- Property descriptions
- Real estate blogs and documentation

The URL loader acts as a web content extraction tool. It visits the specified webpage, downloads its content, removes irrelevant elements such as advertisements, navigation menus, and page styling, and converts the meaningful textual information into LangChain **Document** objects.

## Process Flow

```text
Real Estate Webpage URL
            │
            ▼
   UnstructuredURLLoader
            │
            ▼
   Content Extraction
            │
            ▼
   LangChain Documents
```

## Loading Documents

When the `loader.load()` method is executed, the webpage content is fetched and stored in the variable `docs`.

```python
docs = loader.load()
```

The output is a list of Document objects.

### Structure of a Document Object

Each document consists of two primary components:

#### 1. Page Content

Contains the actual text extracted from the webpage.

```python
docs[0].page_content
```

Example:

```text
Property Name
Location
Price
Amenities
Property Description
Builder Information
```

This textual content becomes the knowledge source for the RAG system.

---

#### 2. Metadata

Contains information about the source document.

```python
docs[0].metadata
```

Example:

```python
{
    'source': 'https://example-property-site.com'
}
```

Metadata helps maintain traceability by identifying where the information originated.

---

## Document Count

The total number of loaded documents can be checked using:

```python
len(docs)
```

If a single URL is provided, the result is typically:

```python
1
```

If multiple URLs are supplied, the number corresponds to the total documents loaded.

---

## Role in the RAG Pipeline

The extracted document content serves as the foundation of the entire RAG workflow.

```text
Document Loading
        │
        ▼
Text Splitting
        │
        ▼
Embedding Generation
        │
        ▼
Vector Database Storage
        │
        ▼
Similarity Retrieval
        │
        ▼
LLM Response Generation
```

Without document loading, the assistant would have no external knowledge source to retrieve information from.

# 📊 Step 2: Loading Property Data using CSVLoader

In many real-world Real Estate applications, property information is often stored in structured formats such as CSV files. These files may contain details such as property names, locations, prices, amenities, property types, and descriptions. To load this structured data into the RAG pipeline, LangChain provides the **CSVLoader**.

The `CSVLoader` is imported from LangChain's document loader module and initialized by specifying the path to the CSV file using the `file_path` parameter.

## Process Flow

```text
Property Listings CSV File
            │
            ▼
          CSVLoader
            │
            ▼
     Row-by-Row Processing
            │
            ▼
   LangChain Document Objects
```

## Loading the CSV File

When the `loader.load()` method is executed, the CSV file is opened and processed row by row.

```python
docs = loader.load()
```

Each row in the CSV file is automatically converted into a separate LangChain **Document** object. The resulting documents are stored in the variable `docs`, which is a list of Document objects.

---

## Number of Documents Created

The total number of generated documents can be determined using:

```python
len(docs)
```

Since each row becomes a separate Document, the number of documents is generally equal to the number of records present in the CSV file.

For example:

```text
100 Property Records
        ↓
100 LangChain Documents
```

---

## Understanding a Document Object

The following command displays the first document:

```python
docs[0]
```

Each LangChain Document consists of two major components:

### 1. Page Content

The `page_content` attribute contains the textual representation of the data stored in that row.

```python
docs[0].page_content
```

Example:

```text
Property Name: Green Valley Apartments
Location: Bangalore
Price: ₹85 Lakhs
Bedrooms: 3
Amenities: Gym, Swimming Pool, Parking
```

This textual content becomes the knowledge source used by the RAG system.

---

### 2. Metadata

The `metadata` attribute contains additional information about the document.

```python
docs[0].metadata
```

Example:

```python
{
    'source': 'property_listings.csv',
    'row': 0
}
```

Metadata helps identify the origin of information and enables traceability throughout the retrieval process.

---

## Viewing the Extracted Content

The actual content of the first document can be displayed using:

```python
print(docs[0].page_content)
```

This output represents the textual version of the first property record in the dataset.

Once the CSV data has been converted into Document objects, it becomes part of the Retrieval-Augmented Generation workflow.

```text
CSV File

   ▼
CSVLoader
   │
   ▼
Document Objects
   │
```

The loaded property records serve as the knowledge base from which relevant information can later be retrieved.

## Application in the Real Estate Assistant

In this project, the CSV file acts as a structured property database containing real estate information. Each property listing is converted into a separate Document, making it searchable through semantic similarity techniques.

Typical information stored in the CSV file may include:

* Property Name
* Property Type
* Location
* Price
* Number of Bedrooms
* Area (Sq. Ft.)
* Amenities
* Property Description
* Builder Information

After loading, these documents are converted into embeddings and stored inside a vector database such as ChromaDB. When users ask questions, the system retrieves the most relevant property records and uses them as context for generating accurate responses.

### Example User Queries

* Show properties available in Bangalore under ₹1 Crore.
* Which apartments provide a swimming pool and gym?
* Compare two properties based on price and amenities.
* What are the details of Property X?

The CSVLoader therefore serves as an essential component for importing structured real estate data into the RAG pipeline, enabling efficient semantic search and intelligent question answering.
