### Manual approach of splitting the text into chunks

```python
text[0:100]
```

```python
# Well but we want complete words and want to do this for entire text, may be we can use Python's split funciton

words = text.split(" ")
len(words)
```

```python
chunks = []

s = ""
for word in words:
    s += word + " "
    if len(s)>200:
        chunks.append(s)
        s = ""
        
chunks.append(s)
```

```python
chunks[:2]
```

Splitting data into chunks can be done in native python but it is a tidious process. Also if necessary, you may need to experiment with various delimiters in an iterative manner to ensure that each chunk does not exceed the token length limit of the respective LLM.

Langchain provides a better way through text splitter classes.

# ✂️ Manual Text Chunking

This code demonstrates a basic text chunking technique used in RAG systems. The document is first stored in the variable `text` and then split into individual words using `split(" ")`.

A loop continuously adds words to a temporary string until its length exceeds 200 characters. The chunk is then stored in the `chunks` list, and a new chunk is created. After processing all words, any remaining text is added to the final chunk.

The generated chunks can be viewed using `chunks[:2]`.

### Why Chunking?

Large documents cannot be directly stored or retrieved efficiently. Therefore, they are divided into smaller chunks before:

* Embedding Generation
* Vector Database Storage
* Semantic Retrieval

This is a simple chunking method for learning purposes. In production RAG systems, tools such as **RecursiveCharacterTextSplitter** are preferred because they preserve context and support chunk overlap.

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size=200,
    chunk_overlap=0
)
```

```python
chunks = splitter.split_text(text)
len(chunks)
```

As you can see, all though we gave 200 as a chunk size since the split was based on \n, it ended up creating chunks that are bigger than size 200.

# ✂️ CharacterTextSplitter

This code uses LangChain's **CharacterTextSplitter** to divide a large document into smaller chunks. The splitter is configured with `separator="\n"`, which uses newline characters as preferred splitting points, `chunk_size=200` to limit each chunk to approximately 200 characters, and `chunk_overlap=0` so that no content is repeated between chunks.

The `split_text(text)` method processes the input text and returns a list of chunks stored in the variable `chunks`.

The expressions `len(chunks)` and `len(chunks[0])` help verify the number of chunks created and their sizes.

### Why is this Important?

In a RAG system, large documents are split into smaller chunks before:

* Embedding Generation
* Vector Database Storage
* Semantic Retrieval

Smaller chunks improve retrieval accuracy by helping the system locate relevant information more precisely during similarity search.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", " "],  # List of separators based on requirement (defaults to ["\n\n", "\n", " "])
    chunk_size = 200,  # size of each chunk created
    chunk_overlap  = 30,  # size of  overlap between chunks in order to maintain the context
    length_function = len  # Function to calculate size, currently we are using "len" which denotes length of string however you can pass any token counter)
)
```

```python
chunks = r_splitter.split_text(text)

for chunk in chunks:
    print(len(chunk))
```

```python
print(chunks[0])
```

```python
print(chunks[1])
```

As you see above, the second chunk has an overlap of few end characters from the first chunk due to chunk_overlap

#### Let's understand how exactly it formed these chunks

```python
first_split = text.split("\n\n")[0]
first_split
```

```python
len(first_split)
```

Recursive text splitter uses a list of separators, i.e. separators = ["\n\n", "\n", "."]

So now it will first split using \n\n and then if the resulting chunk size is greater than the chunk_size parameter which is 200 in our case, then it will use the next separator which is \n

```python
second_split = first_split.split("\n")
second_split
```

```python
len(second_split[0])
```

```python
third_split = second_split[0].split(" ")
len(third_split)
```

```python
third_split[0]
```

# 🔄 RecursiveCharacterTextSplitter

This code uses LangChain's **RecursiveCharacterTextSplitter**, one of the most widely used text splitters in RAG applications. It splits text intelligently by trying multiple separators in order: paragraphs (`\n\n`), lines (`\n`), and spaces (` `).

The splitter is configured with `chunk_size=200` and `chunk_overlap=30`, which means each chunk can contain up to 200 characters and shares 30 characters with the next chunk to preserve context.

The `split_text(text)` method generates a list of chunks, while `len(chunk)` and `chunks[0]` help verify the chunk sizes and content.

### Why is this Important?

In a RAG system, preserving context is critical for accurate retrieval. RecursiveCharacterTextSplitter creates more meaningful chunks by respecting the document structure and maintaining overlap between chunks.

The generated chunks are then used for:

* Embedding Generation
* Vector Database Storage
* Semantic Search
* Question Answering

This makes retrieval more accurate and reduces the loss of important context.
